```python
# To visualize the correlation between aata values, we need to draw a heatmap.
plt.figure(figsize=(10,6))
sns.heatmap(titanic.corr(numeric_only = True), annot = True, cmap='coolwarm')
plt.show()
```

```python
# To view the columns to work on the graphical presentations of the Data.
titanic.columns
```

```python
# The age distribution of the passengers has been presented in histogram.
plt.figure(figsize=(10, 6))
sns.histplot(x='Age', kde=True, hue='Sex', data=titanic)
plt.title('Age Distribution By Gender')
plt.show()
```

```python
# To view the columns to work on the graphical presentations of the Data.
titanic.columns
```

```python
# Visualizing the survival rates based on passenger class and gender of the passengers.
# Female passenger had a significantly higher survival rate than male passengers. first class passengers had more survival rate than
# two other passenger classes.
plt.figure(figsize=(10, 6))
sns.barplot(data=titanic, x='Pclass', y='Survived', hue='Sex')
plt.title('Survival Rate by Passenger Class and Gender')
plt.show()
```

```python
titanic.columns
```

```python
# To compare the Fare between three Passenger Classes. The Fare for the first class is higher than other two passenger
# classes.
plt.figure(figsize=(10, 6))
sns.barplot(data=titanic, x='Pclass', y='Fare', hue='Pclass')
plt.title('Comparison of Price Between Passenger Class')
plt.show()
```

```python
# To view the columns to work on the graphical presentations of the Data.
titanic.columns
```

```python
# Visualization of passenger survival rates based on embarkation port.
# The survival count varied based on the embarkation port.
plt.figure(figsize=(10, 6))
sns.countplot(data=titanic, x='Embarked', hue='Survived')
plt.title('Survival Count Based on Embarkation Port')
plt.show()
```

```python
# To view the columns to work on the graphical presentations of the Data.
titanic.columns
```

```python
# Visualization of survival based on Fare and Passenger Class.
plt.figure(figsize=(10, 6))
sns.boxplot(data=titanic, x='Pclass', y='Fare', hue='Survived')
plt.ylim(0, 300)  # Limiting y-axis to 300 for better visualization
plt.title('Fare distribution by Passenger Class and Survival')
plt.show()
```

```python
# To view the columns to work on the graphical presentations of the Data.
titanic.columns
```

```python
# A new column Family Size has been created by adding 'SibSp', 'Parch' columns of the Dataset.
titanic['Family_Size'] = titanic['SibSp'] + titanic['Parch']
```

```python
# Visualization of the survival count based on the family size (Calculted based on 'SibSp', 'Parch' columns of the Dataset). 
# Solo travelers and those with a small family size (1-3 members) had better survival outcomes than larger families.
plt.figure(figsize=(10, 6))
sns.countplot(data=titanic, x='Family_Size', hue='Survived')
plt.title('Survival Count based on Family Size')
plt.show()
```

```python
# To view the columns to work on the graphical presentations of the Data.
titanic.columns
```

```python
# Visualization of the survival rates by the number of siblings/spouses aboard.
# Those with one sibling or spouse onboard seemed to have a slightly better survival rate than those alone or with many siblings/spouses.
plt.figure(figsize=(10, 6))
sns.countplot(data=titanic, x='SibSp', hue='Survived')
plt.title('Survival Count Based on Number of Siblings/Spouses Aboard')
plt.show()
```
